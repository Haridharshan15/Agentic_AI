#Custom tools (method 1)

```python
from langchain_core.tools import tool
```

```python
#step 1--create a funtion
def multiply(a,b):
  """Multiply two numbers"""
  return a*b
```

```python
#step 2--addind type hints
def multiply(a: int,b: int)->int:
  return a*b
```

```python
#step 3 -- adding tool decorater
@tool
def multiply(a: int,b: int)->int:
  "Multiplies two integers and returns the result."
  return a*b
```

```python
multiply(5,3)
```

```python
def multiply1(a,b):
  """Multiply two numbers"""
  return a*b
```

```python
multiply1(5,3)
```

```python
output = multiply.invoke({"a":5,"b":3})
print(output)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```
