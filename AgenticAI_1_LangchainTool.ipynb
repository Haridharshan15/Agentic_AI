#Custom tools (method 1) using @tool to create our custom tool


```python
from langchain_core.tools import tool
```

```python
#step 1--create a funtion
def multiply(a,b):
  """Multiply two numbers"""
  return a*b
```

```python
#step 2--addind type hints
def multiply(a: int,b: int)->int:
  return a*b
```

```python
#step 3 -- adding tool decorater
@tool
def multiply(a: int,b: int)->int:
  "Multiplies two integers and returns the result."
  return a*b
```

```python
multiply(5,3)
```

```python
def multiply1(a,b):
  """Multiply two numbers"""
  return a*b
```

```python
multiply1(5,3)
```

```python
output = multiply.invoke({"a":5,"b":3})
print(output)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
multiply.args_schema.model_json_schema()#LLM expecting from the tool response like this output ,Json format
"""Note: using LLM we are actually calling a tool,
that tool has a specific funtion"""
```

#Custom tools(Method 2)

StructuredTool & Pydantic class

```python
from langchain_core.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
#arg schema using pyndantic
class MutiplyInput(BaseModel):
    a:int = Field(required=True,description="first number")
    b:int = Field(required=True,description="second number")
```

```python
def multiply_func(a: int,b: int)-> int:
  return a*b
```

```python
multiply_tool = StructuredTool.from_function(
    func = multiply_func,
    name = "multiply",
    description = "Multiply two numbers",
    args_schema = MutiplyInput
)
```

```python
result = multiply_tool.invoke({'a':3,'b':5})
print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

#Custom tools(Method 3)

Basetool

```python
from langchain_core.tools import BaseTool
from typing import Type
```

```python
class MutiplyInput(BaseTool):
    a:int = Field(required=True,description="first number")
    b:int = Field(required=True,description="second number")
```

```python
class MultiplyTool(BaseTool):
  name : str= "multiply"
  description:str = "Multiply two numbers"
  args_schema: Type[MutiplyInput] = MutiplyInput

  def _run(self, a: int, b: int) -> int:
    return a * b
# def _run(self,a,b):
 #   return a*b
```

```python
multiply_tools = MultiplyTool()
```

```python
result = multiply_tools.invoke({'a':3,'b':9})
print(result)
print(multiply_tools.name)
print(multiply_tools.description)
print(multiply_tools.args)
```
